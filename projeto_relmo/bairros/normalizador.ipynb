```python
def retorna_normalizados(numeros):
    #if max(numeros != 0.0)
    normalizados = []
    deno = float(max(numeros)) - float(min(numeros))
    print("deno", deno)
    print("Min",min(numeros ))
    print("Max",max(numeros) )
          
    for item in numeros:
        print("item: ", item)
        item_norm_1 = float(item) - min(numeros)
        print("menos", item_norm_1)
        item_norm = item_norm_1 / deno
        print("divisao", item_norm)
        normalizados.append(item_norm)
    #if len(normalizados) == 75:
    return np.array(normalizados)
    #else:
    #    print("deu ruim aqui, patrao")
```

```python
colunas = ['Bairros', 'agua', 'alfabetizacao', 'coleta_lixo', 'morte_infantil', 'tratamento_sanitario', 'renda_media',  'mais_10_salarios',  'menos_1_salario']
```

```python
agua_norm = retorna_normalizados(data['agua'].values)
alfa_norm = retorna_normalizados(data['alfabetizacao'].values)
lixo_norm = retorna_normalizados(data['coleta_lixo'].values)
morte_norm = retorna_normalizados(data['morte_infantil'].values)
trat_norm = retorna_normalizados(data['tratamento_sanitario'].values)
renda_norm = retorna_normalizados(data['renda_media'].values)
dez_norm = retorna_normalizados(data['mais_10_salarios'].values)
um_norm = retorna_normalizados(data['menos_1_salario'].values)

    
```

```python
dados = {'Bairros': data['Bairros'].values,
         'agua' : agua_norm,
        'alfabetizacao' : alfa_norm,
        'coleta_lixo' : lixo_norm,
        'morte_infantil' : morte_norm,
        'tratamento_sanitario' : trat_norm,
        'renda_media' : renda_norm,
        'mais_10_salarios' : dez_norm,
        'menos_1_salario' : um_norm}
```

```python
df = pd.DataFrame(dados, columns = colunas)
df.to_csv("dados_normalizados.csv")
```
