```
data2 = data.drop(index=index_lixos, axis=0)
data2
```

```
#ler customers

print("Lendo customers...\n")
i = 0
customers = []

for item in data2['CustomerID'].values:
    if item not in customers:
        customers.append(item)
    i += 1
```

```
#fazer a matriz dos recibos

#ler products

print("Lendo recibos...\n")
i = 0
recibos = []

for item in data2['InvoiceNo'].values:
    if item not in recibos:
        recibos.append(item)
    i += 1
```

```
clientes_recibos = []
for item in recibos:
  #cluster = cluster_output[cluster_output['produtos'] == produto]['cluster'].values
  cliente = data2[data2['InvoiceNo'] == item]['CustomerID'].values
  clientes_recibos.append(cliente[0])
```

```
df_cliente_recibos = pd.DataFrame(index = recibos,data=clientes_recibos)
```

```
#criar um df com ticket medio
#ticket médio: soma do que gastou/numero de recibos

total_gasto = []
for item in recibos:
  #cluster = cluster_output[cluster_output['produtos'] == produto]['cluster'].values
  preco = data2[data2['InvoiceNo'] == item]['UnitPrice'].values
  quantidade = data2[data2['InvoiceNo'] == item]['Quantity'].values
  #print(preco)
  #print(quantidade)
  gasto = np.dot(preco, quantidade)
  #print(gasto)
  total_gasto.append(gasto)
len(total_gasto)
```

```
df_cliente_recibos['gasto'] = total_gasto
```

```
df_cliente_recibos.columns = ['cliente', 'gasto_neste_recibo']
```

```
z = 0
gasto_corrigido_inflacao = []
while z < len(df_cliente_recibos):
  if (df_cliente_recibos['data_arrumada'][z].strftime("%Y") == "2011"):
    if (df_cliente_recibos['data_arrumada'][z].strftime("%m") != "12"):
      gasto = df_cliente_recibos['gasto_neste_recibo'][z] * ((1 + i_mes)** (12 - int(df_cliente_recibos['data_arrumada'][z].strftime("%m"))))
      gasto_corrigido_inflacao.append(round(gasto,2))
    elif (df_cliente_recibos['data_arrumada'][z].strftime("%m") == "12"):
      gasto = df_cliente_recibos['gasto_neste_recibo'][z]
      gasto_corrigido_inflacao.append(round(gasto,2))
  if (df_cliente_recibos['data_arrumada'][z].strftime("%Y") == "2010"):
    gasto = df_cliente_recibos['gasto_neste_recibo'][z] * ((1 + i_mes)** 12)
    gasto_corrigido_inflacao.append(round(gasto,2))
    #df_cliente_recibos['gasto_neste_recibo'][z] + (i_mes)**(12 - df_cliente_recibos['data_arrumada'][z].strftime("%b"))
  z += 1
len(gasto_corrigido_inflacao)
```
