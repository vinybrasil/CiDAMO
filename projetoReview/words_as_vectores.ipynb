```python
import pandas as pd
from textblob import TextBlob
import os
from unidecode import unidecode
import warnings
import re
import string


warnings.filterwarnings('ignore')
```

```python
print(os.getcwd())
data = pd.read_csv("final.csv")
data = data.drop('Unnamed: 0', axis=1)
data.head()
```

```python
data.describe()
```

```python
data['review_rate'].value_counts()
```

```python
def clean_text_round1(text):
    #remove pontuacao, palavras com numeros, deixa o texto em caixa baixa e remove o texto em []
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = unidecode(text)
    return text

round1 = lambda x: clean_text_round1(x)
```

```python
data_clean = pd.DataFrame(data.review_body.apply(round1))
data_clean
```

```python
stop_words = []
for line in open("stopwords",encoding='utf-8'):
    line = line.rstrip("\n")
    stop_words.append(line)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(encoding='utf-8', stop_words=stop_words)
data_cv = cv.fit_transform(data_clean.review_body)
data_dtm = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names())
data_dtm
```

```python
#data_dtm['parabéns'].sum()
palavras_usadas = {}
for column in data_dtm.columns:
    #print(column,":", data_dtm[column].sum())
    palavras_usadas[column] = data_dtm[column].sum()
```

```python
print("30 palavras que mais aparecem:\n")
for best_positive in sorted(
    palavras_usadas.items(), 
    key=lambda x: x[1], 
    reverse=True)[:30]:
    print (best_positive)

```
