```
#fazer a matriz dos recibos

#ler products

print("Lendo recibos...\n")
i = 0
recibos = []

for item in data['InvoiceNo'].values:
    if item not in recibos:
        recibos.append(item)
    i += 1
```

```
clientes_recibos = []
for item in recibos:
  #cluster = cluster_output[cluster_output['produtos'] == produto]['cluster'].values
  cliente = data[data['InvoiceNo'] == item]['CustomerID'].values
  clientes_recibos.append(cliente[0])
```

```
#criar um df com ticket medio
#ticket mÃ©dio: soma do que gastou/numero de recibos

total_gasto = []
for item in recibos:
  #cluster = cluster_output[cluster_output['produtos'] == produto]['cluster'].values
  preco = data[data['InvoiceNo'] == item]['UnitPrice'].values
  quantidade = data[data['InvoiceNo'] == item]['Quantity'].values
  #print(preco)
  #print(quantidade)
  gasto = np.dot(preco, quantidade)
  #print(gasto)
  total_gasto.append(gasto)
len(total_gasto)
```

```
clientes_recibos
```

```
df_cliente_recibos_total = pd.DataFrame(index = recibos,data=(clientes_recibos))
```

```
df_cliente_recibos_total['gasto'] = total_gasto
```

```
df_cliente_recibos_total.columns = ['cliente', 'gasto_neste_recibo']
```
