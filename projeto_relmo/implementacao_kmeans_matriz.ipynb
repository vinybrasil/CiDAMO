```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("matriz_de_quantidades.csv", index_col=0)
```

```python
data
```

```python
data.sample(15)
```

```python
arr = data.to_numpy()
```

```python
print(arr)
```

```python
erro = []
for i in range(1, 11):
    #kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=2)
    kmeans = KMeans(n_clusters=i, max_iter=300, n_init=10, random_state=2)
    kmeans.fit(arr)
    erro.append(kmeans.inertia_)
plt.plot(range(1, 11), erro)
plt.title('Método do cotovelo')
plt.xlabel('Número de clusters')
plt.ylabel('SSE')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0) #TA LEVANDO EM CONTA O INDEX
```

```python
pred_y = kmeans.fit_predict(arr)
```
