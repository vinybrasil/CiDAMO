```
import numpy as np
import pandas as pd

data = pd.read_csv("data.csv", encoding='latin1') 
data2 = data.drop(['InvoiceNo', 'StockCode', 'InvoiceDate', 'UnitPrice'], axis=1)
data3 = data2[:20000]
```

```
print("Lendo customers...\n")
i = 0
customers = []

for item in data3['CustomerID'].values:
    if item not in customers:
        customers.append(item)
    print("Analisando o ", i, " cliente de ", len(data3['CustomerID'].values))
    i += 1

print("\nLeitura dos customers finalizada.\nLendo os produtos...\n")
```

```
j = 0
products = []
for product in data3['Description'].values:
    if product not in products:
        products.append(product)
    print("Analisando o ", j, " produto de ", len(data3['Description'].values))
    j += 1
```

```
total = []
n = 0
print("\nAnalisando as quantidades de cada cliente...\n ")
for client in customers:
    print("Analisando o ", n, "cliente de ", len(customers))
    quantidades = []
    for product in products:
        x = data[(data['CustomerID'] == client) & (data['Description'] == product)]['Quantity'].values
        if (len(x) == 0):
            quantidades.append(0)
        elif (len(x) == 1):
            quantidades.append(x[0])
        elif (len(x) > 1):
            quantidades.append(sum(x))
        else:
            print("erro")

    total.append(quantidades)
    n += 1
```

```
d = np.array(total)
print(d.shape)
df = pd.DataFrame(d, index=customers, columns=products)
```
